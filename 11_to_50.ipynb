```python
# 14 prime numbers from 1 to 100
flag=False
for i in range(3,100):
    if i%2==0:
        print(i,"is not a prime number.")
    else:
        for j in range(3,i):
            if i%j==0:
                flag=True
                break
    if flag:
        print(i,"is not a prime number.")
    else:
        print(i,"is a prime number.")
```
