```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt 
vec=np.array([[-4],[1]])
origin=np.zeros(vec.shape)#origin point
plt.figure(figsize=(6,6))
plt.quiver(*origin,*vec,color="green",scale=1,units='xy')

plt.grid()
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.gca().set_aspect("equal")
plt.show()

vec
```
