```python
#  employeec class  and a method

class emp():
    name=""
    position=""
    salary=""
    anumsalary=1
    
    def __init__(self,name,position,salary):
        self.name=name
        self.position=position
        self.salary=salary
         
    def ann_sal(self) :
        self.anumsalary=self.salary*12
        return self.anumsalary

emp1=emp("nick","manager",12000)  

print(emp1.ann_sal())    
```
